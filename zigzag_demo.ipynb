Using ZigZag
===
[ZigZag](https://pypi.python.org/pypi/ZigZag) is a (very) small library [I](http://blog.johnbnelson.com) wrote for calculating the peaks and valleys of a sequence (e.g. time series data). It also can calculate the [maximum drawdown](http://en.wikipedia.org/wiki/Drawdown_(economics)), a useful metric for trading analysis. The repository is on github at [https://github.com/jbn/ZigZag](https://github.com/jbn/ZigZag). If you have [numba](http://numba.pydata.org/) installed, it will be very fast; if you don't, it still works fine for most cases. 

This notebook demonstrates how to use ZigZag, and draws attention to a few caveats.

Basic Usage
---

```python
import pandas as pd
from zigzag import peak_valley_pivots, max_drawdown, compute_segment_returns, pivots_to_modes
```

```python
# This is not nessessary to use zigzag. It's only here so that
# this example is reproducible.
np.random.seed(1997)
```

```python
X = np.cumprod(1 + np.random.randn(100) * 0.01)
pivots = peak_valley_pivots(X, 0.03, -0.03)
```

```python
def plot_pivots(X, pivots):
    xlim(0, len(X))
    ylim(X.min()*0.99, X.max()*1.01)
    plot(np.arange(len(X)), X, 'k:', alpha=0.5)
    plot(np.arange(len(X))[pivots != 0], X[pivots != 0], 'k-')
    scatter(np.arange(len(X))[pivots == 1], X[pivots == 1], color='g')
    scatter(np.arange(len(X))[pivots == -1], X[pivots == -1], color='r')
```

The following plot illustrates how the sequence was annotated. 

```python
plot_pivots(X, pivots)
```

The following shows how you can use `pivots_to_modes` to inspect the segments.

```python
pd.Series(X).pct_change().groupby(pivots_to_modes(pivots)).describe().unstack()
```

Calculate the peak to valley returns for all of the segments.

```python
compute_segment_returns(X, pivots)
```

Finally, compute the oft-quoted (in financial literature) `max_drawdown`.

```python
max_drawdown(X)
```

Pandas Compatability
---
The `peak_valley_pivots` function works on pandas `series` assuming the index is either a DateTimeIndex or is \[0, n). [Pandas](http://pandas.pydata.org/) is great.

```python
import pandas as pd
import pandas.io.data as web
```

```python
X = web.get_data_yahoo('GOOG')['Adj Close']
pivots = peak_valley_pivots(X, 0.2, -0.2)
ts_pivots = pd.Series(X, index=X.index)
ts_pivots = ts_pivots[pivots != 0]
X.plot()
ts_pivots.plot(style='g-o')
```
